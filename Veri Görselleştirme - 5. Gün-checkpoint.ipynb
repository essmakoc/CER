```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

#söz konusu verilerin histogramı
n, bins, patches = plt.hist(x, 100, density = 1, facecolor = "b", alpha = 0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor = "g", alpha = 0.25)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogrami: $\mu_1 = {},\ \mu_2 = {}$".format(mu1, mu2))
plt.annotate("Gauss eğrisinin tepesi", xy = (110, 0.04), xytext = (130, 0.045),
            arrowprops = dict(facecolor = "black", shrink = 0.1))

plt.text(75, 0.025, r"$\mu_1 = {},\ \sigma_1 = {}$".format(mu1, sigma1))


plt.text(120,0.035, r"$mu_2 = {},\ \sigma_2 = {}$".format(mu2, sigma2))

plt.axis([40, 160, 0.00, 0.05])
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
```

```python
veri.head()
```

```python
veri.info()
```

### Saçılım Grafiği
    - Saçılım Grafiği, iki farklı değişkenin arasındaki ilişkiyi belirlemek için kullanılır.

```python
def sacilimGrafigiCiz():
    sns.scatterplot(x = "boy", y = "kilo", data = veri)
    plt.title("Boy ve Ağırlık İlişki - Beyaz Izgara Tema")
    plt.show()
    
sacilimGrafigiCiz()
```

```python
sns.set_style("white")
sacilimGrafigiCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafigiCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafigiCiz()
```

### Saçılım Grafiği - Kategorik

```python
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", data = veri)
plt.title("Cinsiyete Göre Boy ve Ağırlık Dağılımı")
plt.xlabel("Sporcunun Boyu")
plt.ylabel("Sporcunun Kilosu")

plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet",style = "madalya", data = veri)
plt.title("Madalyalı Sporcuların Boy ve Ağırlık Dağılımı")
plt.xlabel("Sporcunun Boyu")
plt.ylabel("Sporcunun Kilosu")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", 
                style = "madalya", size = "yas", sizes = (15, 200), data = veri)
plt.title("Madalyalı Sporcuların Boy ve Ağırlık Dağılımı")
plt.xlabel("Sporcunun Boyu")
plt.ylabel("Sporcunun Kilosu")
plt.show()
```

### Saçılım Grafiği - Doğrusal Regrasyon

```python
sns.regplot(x = "boy", y = "kilo", data = veri, marker = "+", scatter_kws = {"alpha" :0.2})
plt.title(" Boy ve Ağırlık Dağılımı")
plt.show()
```

    - Doğrusal regrasyon oluşan, doğrunun eğiminin pozitif olmasından anlaşılacağı üzere boy ve kilo arasındaki pozitif ilişki vardır
